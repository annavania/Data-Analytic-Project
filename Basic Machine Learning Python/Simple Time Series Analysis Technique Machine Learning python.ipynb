```python
df = pd.read_csv(r'C:\all dataset\UMTMVS.csv', index_col='DATE', parse_dates=True)
df.index #example of loading data in single code
```

```python
df.loc['2000-01-01':'2015-01-01'] #get all the data from 2000-01-01 till 2015-05-01
```

```python
df.loc['1992-01-01':'2000-01-01':12] #the data every 12 months, from 1992 till 2000 because it's in month format
```

```python
df.resample(rule='AS').mean().head() #calculate the mean value of shipment at the start of every year by calling resample at rule='AS' for Year Start and then calling the aggregate function mean on it.
```

```python
df.rolling(window=10).mean().head(20) # head to see first 20 values 
```

```python
df.rolling(window=30).max()[30:].head(20) # heck out the maximum value from a window of 30 days
```

```python
df.plot()#basic line plot
```

```python
ax = df.plot()
ax.set(title='Value of Manufacturers Shipments', ylabel='Value') #adding title and y label
```

```python
ax = df.plot(figsize=(12,6))
ax.set(title='Value of Manufacturers Shipments', ylabel='Value') # change the plot size via figsize parameter in .plot.
```

```python
ax = df.resample(rule='AS').mean().plot(figsize=(12,6))
ax.set(title='Average of Manufacturers Shipments', ylabel='Value of Mean of Starting of Year')
#Plot the mean of the starting value of every year via calling .plot after resampling with the rule â€˜ASâ€™ as â€˜ASâ€™ is the rule for the starting of the year.
```

```python
ax = df.resample(rule='AS').mean().plot.bar(figsize=(12,6))
ax.set(title='Average of Manufacturers Shipments', ylabel='Value of Mean of Starting of Year');
#bar plot for the mean of starting of every year by calling .bar on top of .plot.
```

```python
ax = df['UMTMVS'].resample(rule='MS').mean().plot(figsize=(15,8), label='Resample MS')
ax.autoscale(tight=True)
df.rolling(window=30).mean()['UMTMVS'].plot(label='Rolling window=30')

ax.set(ylabel='Value of Mean of Starting of Month',title='Average of Manufacturers Shipments')
ax.legend()

#plot the rolling mean and normal mean for the starting of the month
```

```python
ax = df['UMTMVS'].resample(rule='AS').max().plot(xlim=["1999-01-01","2014-01-01"],ylim=[280000,540000], figsize=(12,7))
ax.yaxis.grid(True)
ax.xaxis.grid(True)

#plot the maximum values for the start of every year from 1995 till 2014.
```
